## **Exploration**

This first notebook explores three datasets called: "Commonsense" (cm_ambig.csv), "Deontology" (deontology_test_hard.csv) and "Justice" (justice_test_hard.csv). The objective is to examine how large language models (Mistral and Gemma3 from Ollama) process diverse formats of ethical scenarios. The difference among these datasets lies in their length and structure; while the "Commonsense" dataset is utilized to evaluate how models analyze long, unstructured narratives, the "Deontology" and "Justice" datasets consist of shorter and more structured sentences.

The "Commonsense" data details social dilemmas, like personal relationship conflicts from online forums, requiring the model to analize complex nuances over an extended text body. On the other hand, the "Deontology" and "Justice" files present direct scenarios paired with a categorical binary variable called "label" to indicate the moral of an action. More in detail, "Deontology" focuses on daily duties and excuses, like questioning whether one should make the bed, while "Justice" evaluates fairness in decision-making, like a teacher skipping a scheduled exam, providing a benchmark for model evaluation.

For a better underestanding, is created a variable "word_count" to select just a proper sample from the 995 rows in total. 

```python
df = pd.read_csv("../data/commonsense/cm_ambig.csv", header=None, names=["text"])
df["word_count"] = df["text"].str.split().str.len()
shortest_5 = df.sort_values("word_count").head()
shortest_5[["word_count"]]
```

```python
df["word_count"].describe()
```

After looking the characteristics of "word_count" is decided to look only a range from 90 to 100 words.

```python
sample_df = df[(df["word_count"] >= 90) & (df["word_count"] <= 100)]
sample_3 = sample_df.sample(n=3, random_state=27)
sample_3[["word_count", "text"]]
```

The raw text is splited into a title and their respective story, and removed extra espaces. Furthermore, for further analysis is selected only a sample of 1 row to reduce computational time execution.

```python
commonsense_df = pd.read_csv("../data/commonsense/cm_ambig.csv", header=None, names=["raw_text"])
```

```python
commonsense_df[["title", "story"]] = commonsense_df["raw_text"].str.split("|", n=1, expand=True)
commonsense_df["title"] = commonsense_df["title"].str.strip()
commonsense_df["story"] = commonsense_df["story"].str.strip()
commonsense_subset = commonsense_df.loc[[331]]
commonsense_subset = commonsense_subset[["title", "story"]]
commonsense_subset
```

For a better understanding it was replaced the terms AITA and WIBTA for: 'Am I the asshole' and 'Would I be the asshole', respectively.

```python
replacing = {'AITA': 'Am I the asshole', 'WIBTA': 'Would I be the asshole'}

commonsense_subset['title'] = commonsense_subset['title'].replace(replacing, regex=True)
commonsense_subset['story'] = commonsense_subset['story'].replace(replacing, regex=True)
```

```python
commonsense_subset
```

```python
commonsense_subset.to_csv("../data/samples/commonsense_subset.csv", index=False)
```

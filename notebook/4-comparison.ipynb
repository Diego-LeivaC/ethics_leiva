## **Comparison**

This last notebook compares the performance of the two LLMs on the three ETHICS datasets. The comparison is based on the metrics previously computed in the individual notebooks.

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
mistral = pd.read_csv("../results/mistral_metrics.csv", index_col=0)
gemma = pd.read_csv("../results/gemma_metrics.csv", index_col=0)
```

```python
comparison = pd.concat({"Mistral": mistral, "Gemma3": gemma}, axis=1)
comparison
```

```python
comparison.to_csv("../results/final_comparison.csv")
```

```python
accuracy = pd.DataFrame({"Mistral": mistral.loc["accuracy"], "Gemma3": gemma.loc["accuracy"]})
accuracy = accuracy.drop("Commonsense")
accuracy.loc["Average"] = accuracy.mean()
accuracy.plot(kind="bar")

plt.title("Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
time = pd.DataFrame({"Mistral": mistral.loc["execution_time (seconds)"], "Gemma3": gemma.loc["execution_time (seconds)"]})
time.loc["Average"] = time.mean()
time.plot(kind="bar")

plt.title("Execution Time")
plt.ylabel("Seconds")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
